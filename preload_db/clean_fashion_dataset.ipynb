```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/myntra_products_catalog.csv")
df.head()
```

```python
df = df.rename(columns={"Price (INR)" :"Price (USD)"})
df.head(100)
```

```python
df.info()
```

```python
df.drop_duplicates()
```

```python
df['Price (USD)'] = df['Price (USD)'].apply(    lambda x: x / 10 if 100 <= x < 1000 else x / 100 if x >= 1000 else x).round(2)
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
brands_df = pd.DataFrame({
    'BrandID': range(1, df['ProductBrand'].nunique() + 1),
    'BrandName': df['ProductBrand'].drop_duplicates().values
})
```

```python
brands_df.head()
```

```python
df = df.merge(brands_df, left_on='ProductBrand', right_on='BrandName')
df.drop(columns=['BrandName'], inplace=True)
df.head(10)
```

```python
cols = df.columns.tolist()
pb_index = cols.index('ProductBrand')
new_order = cols[:pb_index + 1] + ['BrandID'] + cols[pb_index + 1:-1]

df = df[new_order]
df.head(10)
```

```python
df.drop(["Gender","NumImages","PrimaryColor"],axis=1,inplace=True)
df.head()
```

```python
brands_df.to_csv('brands.csv', index=False)
df.to_csv('products.csv', index=False)
```
